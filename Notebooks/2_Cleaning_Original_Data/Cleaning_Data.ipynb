```python
unpopular = pd.read_csv("55000_artist_track.csv")
unpopular_df = pd.DataFrame(unpopular)
unpopular.head()
```

```python
len(unpopular)
```

```python
unpopular_unique = unpopular_df.drop_duplicates()
unpopular_table = Table.from_df(unpopular_unique)
unpopular_table.num_rows
```

```python
unpopular_table = unpopular_table.relabeled(['artist', 'song'], ['Artist', 'Track'])
unpopular_table
```

```python
# unpopular_table_new = unpopular_table.with_columns('Position', [0]*unpopular_table.num_rows)
# unpopular_table_new
```

```python
unpopular_table.to_csv('unpopular_unique_artist_track.csv')
billboard_table.to_csv('billboard_unique_with_positions.csv')
```

```python
combined = billboard_table.to_df().append(unpopular_table.to_df(), ignore_index=True)
combined
```

```python
unique_combined = combined.drop_duplicates(subset=['Artist', 'Track'])
unique_combined
```

```python
column_order = ['Track', 'Artist', 'Position', 'Peak', 'Last', 'Weeks', 'Year', 'Month', 'Week']
combined_unique = unique_combined.reindex(columns=column_order)
combined_unique
```

```python
combined_unique = combined_unique.replace(np.nan, 0)
```

```python
combined_unique.to_csv('combined_unique.csv', index=False)
```
