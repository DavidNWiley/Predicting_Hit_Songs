```python
import wget, time, requests as rq, re, pandas as pd
from datetime import timedelta, date
```

```python
url = 'https://www.billboard.com/charts/hot-100/'

week_date = date(2000,1,1)
this_week = date.today()

clean_pos = []
clean_track = []
clean_artist = []
clean_last = []
clean_peak = []
clean_weeks = []
year = []
month = []
week = [] 

position = [i+1 for i in range(100)]

track_rx = '<span class="chart-element__information__song text--truncate color--primary">(.*?)</span>'
artist_rx = '<span class="chart-element__information__artist text--truncate color--secondary">(.*?)</span>'
last_rx = '<span class="chart-element__meta text--center color--secondary text--last">([-]*\d*)</span>'
peak_rx = '<span class="chart-element__meta text--center color--secondary text--peak">(\d+)</span>'
weeks_rx = '<span class="chart-element__meta text--center color--secondary text--week">(\d+)</span>'



while(week_date <= this_week):
    
    try:
        week_url = url+str(week_date.year)+'-'+str(week_date.month).zfill(2)+'-'+str(week_date.day).zfill(2)

        site = rq.get(week_url)
        site_text = site.text

        track_matches = re.findall(track_rx, site_text)
        artist_matches = re.findall(artist_rx, site_text)
        last_matches = re.findall(last_rx, site_text)
        peak_matches = re.findall(peak_rx, site_text)
        weeks_matches = re.findall(weeks_rx, site_text)

        for m in range(100):
            clean_pos.append(position[m])
            clean_track.append(track_matches[m].replace('&amp;', '&').replace('&#039;', "'"))
            clean_artist.append(artist_matches[m].replace('&amp;', '&').replace('&#039;', "'"))
            clean_last.append(last_matches[m].replace('&amp;', '&').replace('&#039;', "'"))
            clean_peak.append(peak_matches[m].replace('&amp;', '&').replace('&#039;', "'"))
            clean_weeks.append(weeks_matches[m].replace('&amp;', '&').replace('&#039;', "'"))
            year.append(week_date.year)
            month.append(week_date.month)
            week.append(week_date.day)

        week_date += timedelta(7)
        print(str(week_date.year) + '-' + str(week_date.month) + '-' + str(week_date.day) + ' done.')
    except:
        pass
```

```python
week_dict = {'Track': clean_track, 
             'Artist': clean_artist,
             'Position': position,
             'Last': clean_last, 
             'Peak': clean_peak, 
             'Weeks': clean_weeks, 
             'Year': year, 
             'Month': month, 
             'Week': week}

week_df = pd.DataFrame(week_dict)
week_df
week_df.to_csv('2000-01-01_to_2019-10-26.csv', index=False)
```
